```python
import pandas as pd

# Create a simple dataset
data = {
    'Name': ['Asha', 'Bala', 'Chitra', 'Deepak', 'Esha', 'Farhan'],
    'Age': [21, 22, 20, None, 23, 22],
    'Department': ['CS', 'IT', 'CS', 'ECE', 'ECE', None],
    'Marks': [85, 90, 78, 92, None, 88],
    'City': ['Chennai', 'Madurai', 'Chennai', 'Trichy', 'Salem', 'Chennai']
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Display the first few rows
df.head()
```

```python
df
```

```python
# Step 3: Data Preprocessing

print("üîç Checking for missing values...")
print(df.isnull().sum())  # Shows how many missing values each column has

df = df.fillna({
    'Age': df['Age'].mean(),
    'Marks': df['Marks'].mean(),
    'Department': 'Unknown'
})


print("\n‚úÖ Missing values handled successfully!")
print("\nCleaned Data:")
print(df)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Step 4: Data Transformation
print("‚öôÔ∏è Transforming data...")

# 1Ô∏è‚É£ Scale numeric columns
scaler = StandardScaler()
df[['Age', 'Marks']] = scaler.fit_transform(df[['Age', 'Marks']])

# 2Ô∏è‚É£ Encode categorical columns
label_enc = LabelEncoder()
df['Department'] = label_enc.fit_transform(df['Department'])
df['City'] = label_enc.fit_transform(df['City'])

print("\n‚úÖ Data transformation complete!")
print("\nTransformed Data:")
print(df)
```

```python
# Step 5: Load the data

# Save transformed data to CSV
df.to_csv('transformed_data.csv', index=False)

# Provide a download link
from google.colab import files
files.download('transformed_data.csv')

print("‚úÖ Transformed data saved as 'transformed_data.csv' and ready to download!")
```
